Exemplo: Uma padaria produz pães que proporcionam um lucro de R\$1,00 real, quando vendidos, ou um prejuízo de R$0,50 quando são encaminhados para virarem farinha, no outro dia. Se a probabilidade de vender um pão é igual a 0,8, qual o lucro médio da padaria?

```python
import numpy as np
```

```python
Ns = 1000000
p = 0.8
lucro = 1
prejuizo = 0.5
ganho = 0
for s in range(Ns):
    if np.random.uniform() < 0.8:
        ganho = ganho + lucro
    else:
        ganho = ganho - prejuizo

print('Lucro médio: ', ganho/Ns)
```

A interpretação frequentista da probabilidade diz que, se você repetir o experimento (vender pães) inúmeras vezes sob as mesmas condições, a média dos seus resultados vai convergir exatamente para esse valor teórico.
Portanto, "depois de um tempo" (ou após muitas vendas), o seu lucro total acumulado dividido pelo número de pães produzidos será muito próximo de R$ 0,70

**Exemplo:** Lançamos dois dados e observamos a variável aleatória
X , que é igual a 1 se o valor observado na soma de dois dados for
par, ou igual a zero, caso contrário. Calcule a esperança de
g (X ) = X + 2.

Valor calculado na aula: E[g(X)] = 2,5

```python
def g(x):
    x = x + 2
    return x

dado = np.array([1,2,3,4,5,6])
ns = 1000000
vg = []
for i in range(ns):
    x = 0
    dado1 = np.random.choice(dado)
    dado2 = np.random.choice(dado)
    soma = dado1 + dado2

    if(soma % 2 == 0):
        x = 1
    vg.append(g(x))

print('Esperança: ', np.mean(vg))
```

O valor 2,5 indica que, se você repetir esse experimento muitas vezes, a média dos valores gerados pela função g(X) será exatamente 2,5

A utilização de uma função g(X) nesse exemplo serve para calcular o valor esperado de uma transformação ou de uma regra adicional aplicada ao resultado da variável aleatória original X
