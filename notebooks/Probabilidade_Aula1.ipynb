```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

Simular o lançamento de uma moeda e calcular a frequência de ocorrência de caras

```python
moeda = ['cara','coroa']
fc = 0
n = 20
for i in range(0,n):
    saida = random.choice(moeda)
    if(saida == 'cara'):
        fc = fc + 1
    print(saida)
print('Frequencia: ',fc/n)
```

```python
# Variando o número de variações
# Quanto maior o numero de variaçoes ou seja do número de experimentos, mais a probabilidade converte para o teórico

moeda = ['cara',  'coroa']
p = 0.5
lcaras = []
ln = []
for n in range(1,1000):
  ncaras = 0
  for i in range(0,n):
    face = np.random.choice(moeda)
    if face == 'cara':
      ncaras = ncaras + 1
  fa = ncaras/n
  lcaras.append(fa)
  ln.append(n)
plt.plot(ln,lcaras)
plt.axhline(y=p, color='r', linestyle='-')
plt.xlabel('Número de lançamentos')
plt.ylabel('Frequência de caras')
plt.show()
```

Simulando o lançamento de um dado

```python
dado = [1,2,3,4,5,6]
n = 100000
vfaces = np.zeros(6)
for i in range(0,n):
    face = np.random.choice(dado)
    vfaces[face - 1] = vfaces[face - 1] + 1
print('Valor teórico: ',1/6)
print("Frequencias obtidas: ", vfaces/n)
```

Exemplo das fichas numeradas de 1 a 20.
- Qual a probabilidade de que a ficha escolhida contenha um número maior que 9.

```python
n = 20
urna = range(1,n+1)
valor = 9
f = 0
ns = 10000
for i in range (0,ns):
    ficha = random.choice(urna)
    if(ficha> valor):
        f = f + 1
print('Frequencia:',f/ns)
print('Probabilidade exata: ', len(range(valor+1,n+1))/n)
```

Exemplo das urnas com bolas pretas e brancas
- Probabilidade de tirar uma bola preta

```python
urna = ['B']*3000 + ['P']*2000
nsim = 100000
fA = 0
for i in range(0,nsim):
    bola = random.choice(urna)
    if(bola == 'P'):
        fA = fA + 1
fA = fA/nsim
print('Probabilidade teórica: ', 2/5)
print('Frequencia de bolas pretas: ',fA)
```
