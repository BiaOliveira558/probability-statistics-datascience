```python
import random
```

```python
moeda = ['cara','coroa']
for i in range(0,10):
    saida = random.choice(moeda)
    print(saida)
```

```python
n = 20
urna = range(1,n+1)
valor = 9
f = 0
ns = 100
for i in range (0,ns):
    ficha = random.choice(urna)
    if(ficha> valor):
        f = f + 1
print('Frequencia:',f/ns)
print('Probabilidade exata: ', len(range(valor+1,n+1))/n)
```

```python
urna = ['B']*3000 + ['P']*2000
nsim = 100
fA = 0
for i in range(0,nsim):
    bola = random.choice(urna)
    if(bola == 'P'):
        fA = fA + 1
fA = fA/nsim
print('Frequencia de bolas pretas: ',fA)
```
