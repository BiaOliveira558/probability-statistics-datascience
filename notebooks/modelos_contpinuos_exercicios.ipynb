1 - Uma fábrica de carros sabe que os motores de sua fabricação têm duração normal com média de 150.000 km e desvio padrão de 5.000 km. Qual a probabilidade de que um carro, escolhido ao acaso, dos fabricados por essa firma, tenha um motor que dure:
* a) menos de 170.000 km?
* b) entre 140.000 km e 165.000 km?
* c) Se a fábrica substitui o motor que apresenta duração inferior à garantia, qual deve ser esta garantia para que a porcentagem de motores substituídos seja inferior a 0,2\%?

```python
# a)
norm.cdf(4, loc=0, scale=1)
```

```python
z1 = norm.cdf(3, loc=0, scale=1)
z2 = norm.cdf(-2, loc=0, scale=1)
```

```python
# b)
print( z1 - z2)
```

```python
# c)
x = ((-2.88)*5000) + 150000
x
```

```python
# do professor

# ATENÇÃO: Regra para uso da normal no Python
# P(X < x) = norm.cdf(x, mean, std)
# P(X > x) = 1 - P(X < x) = 1 - norm.cdf(x, mean, std)
# P(x1 < X < x2) = P(X < x2) - P(X < x1) = norm.cdf(x2, mean, std) - norm.cdf(x1, mean, std)

from scipy.stats import norm

# média e desvio padrão
mean = 150000
std_dev = 5000

# a) Probabilidade de que o motor dure menos de 170.000 km:
prob_less_170000 = norm.cdf(170000, loc=mean, scale=std_dev)
print(f"a) Probabilidade de durar menos de 170.000 km: {prob_less_170000:.4f}")

# b) Probabilidade de durar entre 140.000 km e 165.000 km:
prob_between = norm.cdf(165000, loc=mean, scale=std_dev) - norm.cdf(140000, loc=mean, scale=std_dev)
print(f"b) Probabilidade de durar entre 140.000 km e 165.000 km: {prob_between:.4f}")

# c) Garantia para que a porcentagem de motores substituídos seja inferior a 0,2%:
# Isso significa que precisamos encontrar o valor x tal que P(X < x) = 0.002
guarantee_value = norm.ppf(0.002, loc=mean, scale=std_dev)
print(f"c) A garantia deve ser de aproximadamente: {guarantee_value:.2f} km")
```

* b) entre R\$ 8.920,00 e R\$ 9.380,00?

```python
# ATENÇÃO: Regra para uso da normal no Python
# P(X < x) = norm.cdf(x, mean, std)
# P(X > x) = 1 - P(X < x) = 1 - norm.cdf(x, mean, std)
# P(x1 < X < x2) = P(X < x2) - P(X < x1) = norm.cdf(x2, mean, std) - norm.cdf(x1, mean, std)

from scipy.stats import norm

# média e desvio padrão
mean_salary = 8000
std_dev_salary = 500

# a) Porcentagem de diretores que recebem menos do que R$ 6,470.00
prob_less_6470 = norm.cdf(6470, loc=mean_salary, scale=std_dev_salary)
percentage_less_6470 = prob_less_6470 * 100
print(f"a) Porcentagem de diretores que recebem menos de R$ 6.470,00: {percentage_less_6470:.2f}%")

# b) Porcentagem de diretores que recebem entre R$ 8,920.00 e R$ 9,380.00
prob_between_8920_9380 = norm.cdf(9380, loc=mean_salary, scale=std_dev_salary) - norm.cdf(8920, loc=mean_salary, scale=std_dev_salary)
percentage_between_8920_9380 = prob_between_8920_9380 * 100
print(f"b) Porcentagem de diretores que recebem entre R$ 8.920,00 e R$ 9.380,00: {percentage_between_8920_9380:.2f}%")
```

```python
from scipy.stats import norm, geom, binom

# Valores atribuídos para a distribuição normal (peso do óleo)
mean_weight = 990
std_dev_weight = 10
rejection_threshold = 976

# Probabilidade de uma única lata ser rejeitada (peso < 976 g)
prob_rejected_single = norm.cdf(rejection_threshold, loc=mean_weight, scale=std_dev_weight)
print(f"Probabilidade de uma única lata ser rejeitada: {prob_rejected_single:.4f}")

# a) Probabilidade de que o 10º observado seja o 1º rejeitado
# Este é um problema de distribuição geométrica
# P(X=k) = (1-p)^(k-1) * p , onde p é a probabilidade de sucesso (rejeição) em uma única tentativa
prob_10th_is_1st_rejected = geom.pmf(10, prob_rejected_single)
print(f"a) Probabilidade da 10ª lata ser a 1ª rejeitada: {prob_10th_is_1st_rejected:.6f}")

# b) Probabilidade de que, em 20 latas observadas, 3 sejam rejeitadas
# Este é um problema de distribuição binomial.
# P(X=k) = C(n, k) * p^k * (1-p)^(n-k)
n_cans = 20
k_rejected = 3
prob_3_out_of_20_rejected = binom.pmf(k_rejected, n_cans, prob_rejected_single)
print(f"b) Probabilidade de, em 20 latas, 3 serem rejeitadas: {prob_3_out_of_20_rejected:.6f}")
```

```python
from scipy.stats import expon

# Média do tempo de espera
mean_wait_time = 4
# A taxa (lambda) da distribuição exponencial é 1/média
lambda_param = 1 / mean_wait_time

# a) Probabilidade de que um cliente espere entre 4 e 5 minutos na fila:
# P(4 < X < 5) = P(X <= 5) - P(X <= 4)
prob_between_4_and_5 = expon.cdf(5, scale=mean_wait_time) - expon.cdf(4, scale=mean_wait_time)
print(f"a) Probabilidade de esperar entre 4 e 5 minutos: {prob_between_4_and_5:.4f}")

# b) Probabilidade de que um cliente espere mais do que 6 minutos na fila:
# P(X > 6) = 1 - P(X <= 6)
prob_more_than_6 = 1 - expon.cdf(6, scale=mean_wait_time)
print(f"b) Probabilidade de esperar mais de 6 minutos: {prob_more_than_6:.4f}")
```

6 - No desenvolvimento de um novo foguete, engenheiros inserem dois tanques de combustível,  sendo um deles ativo e o outro usado como reserva, que é acionado caso ocorra uma falha no primeiro tanque.  Suponha que em uma missão espacial necessita de 50 horas de voo para ser finalizada. De acordo com o fabricante do tanque, o tempo médio antes de ocorrer uma falha é de 100 horas.  Calcule a probabilidade de que a missão será bem-sucedida. Assuma que o tempo de funcionamento dos tanques apresenta distribuição exponencial.

Como temos dois tanques, para que o sistema todo falhe devem ocorer dois eventos (duas falhas), denso que o tempo associado a cada um deles segue o modelo exponencial. Logo, o tempo para o sistema falhar segue uma distribuição gama com a = 0. 

```python
from scipy.stats import expon, gamma

# Tempo médio antes da falha (lambda)
mean_time_to_failure = 100
lambda_param = 1 / mean_time_to_failure

# Tempo necessário para a missão
mission_time = 50

# Com dois tanques, o tempo total de funcionamento é a soma de duas variáveis exponenciais independentes.
# A soma de duas exponenciais independentes com a mesma taxa lambda segue uma distribuição Gamma com parâmetros k=2 e lambda.
# Queremos a probabilidade de que o tempo total de funcionamento (soma dos tempos dos dois tanques) exceda o tempo da missão (50 horas).
# Isso é P(X > 50), onde X segue uma distribuição Gamma com k=2 e lambda=0.01.
# O parâmetro de escala para a distribuição gama no scipy é 1/lambda.

# Probabilidade de que o tempo total exceda o tempo da missão
prob_mission_successful = 1 - gamma.cdf(mission_time, a=2, scale=mean_time_to_failure)

print(f"A probabilidade de que a missão seja bem-sucedida é de: {prob_mission_successful:.4f}")
```
