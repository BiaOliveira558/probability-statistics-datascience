3 - O tempo de vida (em horas) de um componente eletrônico segue aproximadamente uma distribuição normal com
média µ = 500 horas e desvio padrão σ = 50 horas.

(a) Qual a probabilidade de que um componente escolhido ao acaso dure menos de 450 horas?

(b) Qual a probabilidade de que ele dure entre 480 e 520 horas?

(c) Determine o tempo t tal que 90% dos componentes durem mais do que t horas.

(d) Qual é o valor esperado e a variância desta distribuição?

```python
media = 500
dp = 50
k = 450

# 1. P(X <= 450) -> Probabilidade acumulada até um valor
prob_ate_450 = norm.cdf(k, loc=media, scale=dp)
prob_ate_450
```

```python
k1 = 480
k2= 520
media = 500
dp = 50
prob1 = norm.cdf(k1, loc=media, scale=dp)
prob2 = norm.cdf(k2, loc=media, scale=dp)

prob_final = prob2 - prob1
prob_final
```

```python
media = 500
dp = 50
prob_acumulada = 0.10

t = norm.ppf(prob_acumulada, loc=media, scale=dp)
t
```

4 - O tempo (em horas) até a falha de um servidor segue uma distribuição exponencial com taxa de falha λ = 0,05 por
hora.
Responda:

(a) Qual a probabilidade de que o servidor funcione por mais de 20 horas sem falhar?

(b) Qual a probabilidade de que o servidor falhe antes de 5 horas de operação?

(c) Qual o valor esperado e a variância do tempo até a falha?

```python
from scipy.stats import expon

lambd = 0.05
tempo_medio = 1 / lambd

# (a) Probabilidade de funcionar MAIS de 20 horas: P(X > 20)
# Usamos a Survival Function (sf), que calcula 1 - CDF
prob_a = expon.sf(20, scale=tempo_medio)
prob_a
```

```python
# (b) Probabilidade de falhar ANTES de 5 horas: P(X < 5)
prob_b = expon.cdf(5, scale=tempo_medio)
prob_b
```

```python
# (c) Valor esperado (E[X]) e Variância (Var(X))
esperanca = expon.mean(scale=tempo_medio)
variancia = expon.var(scale=tempo_medio)

print(f"(a) P(X > 20): {prob_a:.4f}")
print(f"(b) P(X < 5):  {prob_b:.4f}")
print(f"(c) Média: {esperanca}h, Variância: {variancia}")
```
